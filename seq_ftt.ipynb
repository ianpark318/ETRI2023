```python
import os

import torch
import torch.nn.functional as F

import numpy as np
import pandas as pd
import datetime
from datetime import datetime as dt
```

```python
data_path = "./user01-06/user06"
dir_list = os.listdir(data_path)

user_all_data = []

for item in dir_list:
    sub_path = data_path + "/" + item + "/" + "e4Eda"
    for i in os.listdir(sub_path):
        csv_file = sub_path + "/" + i
        print(csv_file)
        rdr = pd.read_csv(csv_file)
        user_all_data.append(rdr)
        

df_all = pd.concat(user_all_data, axis=0)
df_all['ts']=pd.to_datetime(df_all['ts'], unit='s')
df_all
```

```python
df_all.sort_index()
df_all = df_all.loc[~df_all.index.duplicated(keep='first')]
```

```python
import matplotlib.pyplot as plt


df_all.plot("timestamp", "eda")
plt.show()
```

```python
from scipy.interpolate import splrep, splev


x = df_all["timestamp"]
y = df_all["eda"]

fs = 1000
dt = 1/fs
spl = splrep(x, y)

x1 = np.arange(0, 1, dt)
y1 = splev(x1, spl)

plt.figure(figsize=(15, 5))
plt.plot(x1, y1, marker='o')
plt.show()
```
